## Prerequisites

1. Decision Tree

1. In Adaboost, its just a node and two leaves
2. A tree with one node and two leaves is called stump
3. So it forest of stumps rather than trees
4. Stumps are not great in accurate classifications
5. Stumps are technically weak learner

1. However, that the way Adaboost likes it and its one of the reasons on why they are so commonly combined
2. In Random forest, each tree has the equal weightage of votes
3. While in forest of stumps made with Adaboost, some stumps get more say in the final classification than others
4. In our cases, the larger stumps get more say than the smaller stumps
5. In random forest, each decision tree made independent of other (In other words, it does not matter if which tree is made first)

1. In Forest of stumps, the order is important(the errors made in the first stump affect the second one)
2. Three important concepts of Adaboost are
    - Adaboost combines a lot of weak learners to make classification, the weak learners are almost always stumps
    - Some stumps get more say in the classification of others
    - Each step is made by taking previous stump's mistake into account

## Deep Dive

1. Imagine we have a dataset that predicts if the person is having heart disease or not with the help of the features like chest pain, blocked arteries & weight
2. We are gonna assign sample weight that indicates how important it is to be correctly classified
3. Here the sample weight is different from the patient weight
4. At start all samples get the same weight, (1/total number of samples)
5. After we make the first stump, these weights will change in order to guide how the next stump is created

1. Pick up one of the features and make a stump
2. Repeat the steps for rest of the features as well
3. Calculate the Gini index for all the features
4. The Gini index with the lowest value will the first stump in the forest
5. Now we need to see how much say is this having in the final classification

1. We determine how much say a stump has in the final classification based on how well it classifies the samples
2. The total weight is assiociated with the sum of weights of the incorrect samples
3. The weight with 1 is horrible stump and the weight with 0 is perfect stump
4. Amount of say = 1/2(log(1- total error/Total error))
