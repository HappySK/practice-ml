## Why Random Forest ?

1. The Random Forest is build on the top of decision trees
2. The decision trees are easy to classify and interpret
3. But they only work well with the training data
4. Due to this, it results with inaccuracy when it comes to new data
5. Random forest helps in constructing various trees and vote the top most classification or regression as an output

1. Random forest result in simplicity and improving the vast accuracy
2. Here are the steps
    - Create a Bootstrapped dataset
        - Bootstrapped dataset is something that is a new dataset of same size of the original dataset
        - Here the bootstrapped dataset allows the use of one records in the dataset more than once
    - Create a decision tree from the bootstrapped dataset but with only subset of columns
    - Repeat step 1 and 2 and ideally it becomes 100s of trees
3. The test data is tested on all the 100s of trees and aggregated together
4. The output is determined based on which one has higher votes from the list of trees
5. Bootstrapping the dataset and aggregating the results is known as Bagging

1. Since the bootstrapped dataset contains duplicate dataset, there may be chances that the records have the chances of missing out during the training
2. These kinds of records is known as Out of Bag Dataset
3. Ultimately we can calculate the accuracy of the Random Forest by measuring the proportion of out of bag sample that were correctly classified by the random forest
4. The proportion of out of bag samples that are incorrectly classified is known as Out of Bag error
5. Run on different set of parameters like no of parametes to be selected on the subset of columns

## Dealing with Missing Data

1. Random Forest handles missing data
2. It works on two kinds of missing data
    - Missing data on the independent features
    - Missing data on the dependent feature
3. Missing data on the independent features
    - Get the missing value with guess
    - Refine the guess until we get good accuracy
    - Now we need to find the similarity with the another samples
    - Determining similarities
        - A proximity matrix is determined that tells how similar each values are

## Python Implementation (Classification)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
input = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
input.head()
```

```python
target = pd.DataFrame(data=iris.target, columns=["species"])
```

```python
target.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.predict(X_test)
```

```python
rf.score(X_test, y_test)
```

## Python Implementation (Regression)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import fetch_california_housing
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
california_housing = fetch_california_housing()
```

```python
input = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)
target = pd.DataFrame(data=california_housing.target, columns=["saleprice"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.2)
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```
